```python
def juzzHist ( data, subplot = plt, bintype = 'linear', binNum = 50, orientation = 'verticle', color = 'black', 
              alpha = 0.5, 
              label = '', normalize = True):
    if normalize:
        weights=np.ones(len(data)) / len(data)
    else:
        weights = None
    
    if bintype == 'log':
        bins = np.logspace(np.log10(np.min(data)),np.log10(np.max(data)), binNum)
    elif bintype == 'linear':
        bins = binNum
    else: 
        print('Unrecognized bintype kwarg:', bintype, '. Asserting bintype = "linear"')
    if label != '':
        H = subplot.hist(data, bins = bins, label = label, alpha = alpha, color = color, weights=weights)
        subplot.stairs(H[0],H[1], color = color)
    else:
        H = subplot.hist(data, bins = bins, alpha = alpha, color = color, weights=weights)
        subplot.stairs(H[0],H[1], color = color)
    return H
```
